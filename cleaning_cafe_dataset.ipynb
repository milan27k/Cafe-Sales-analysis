```python
df['Quantity'] = pd.to_numeric(df['Quantity'],errors='coerce')
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'],errors='coerce')
df['Total Spent'] = pd.to_numeric(df['Total Spent'],errors='coerce')
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'],errors='coerce')
```

```python
df.replace("UNKNOWN",np.nan,inplace=True)
```

```python
df.replace("ERROR",np.nan,inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df['Item'] = df['Item'].fillna('Unknown item')
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Quantity'].median())
df['Payment Method'] = df['Payment Method'].fillna(df['Quantity'].mode()[0])
df['Location'] = df['Location'].fillna(df['Location'].mode()[0])
```

```python
df['Total Spent'] = df['Total Spent'].fillna(df['Quantity']*df['Price Per Unit'])
```

```python
df = df.dropna(subset=['Transaction Date'])
```

```python
df.isnull().sum()
```

```python
df['Payment Method'].unique()
```

```python
df['Payment Method'] = df['Payment Method'].replace(["3.0", 3.0],pd.NA)
df['Payment Method'] = df['Payment Method'].fillna(df['Payment Method'].mode()[0])
```

```python
df['Payment Method'].value_counts()
```

```python
df.to_csv("cleaned_cafe_data.csv",index=False)
```
